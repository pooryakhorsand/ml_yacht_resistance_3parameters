# Regression 1 â€“ Deriving Bmax from BWL

Based on the perfect correlation noted in the paper (Section 3.1), we use simple linear regression to derive Bmax from BWL.

```python
# Features and target for Bmax prediction (Bmax ~ BWL)
X_bw = df[['Bwl']]
y_bw = df['Bmax']

# Split the data (75% training, 25% testing)
X_train_bw, X_test_bw, y_train_bw, y_test_bw = train_test_split(X_bw, y_bw, test_size=0.25, random_state=42)

# Train Linear Regression model
model_bw = LinearRegression()
model_bw.fit(X_train_bw, y_train_bw)

# Make predictions
y_pred_bw = model_bw.predict(X_test_bw)

# Evaluate the model
mse_bw = mean_squared_error(y_test_bw, y_pred_bw)
rmse_bw = sqrt(mse_bw)
mae_bw = mean_absolute_error(y_test_bw, y_pred_bw)
r2_bw = r2_score(y_test_bw, y_pred_bw)

# Output regression formula and metrics
print("Regression Formula for Bmax (derived from BWL):")
print(f"Bmax = {model_bw.intercept_:.2f} + ({model_bw.coef_[0]:.2f}) * Bwl\n")
print("Evaluation Results:")
print("MAE:", mae_bw)
print("MSE:", mse_bw)
print("RMSE:", rmse_bw)
print("RÂ² Score:", r2_bw)
```

# Regression 2 â€“ Deriving D from T

Similarly, a linear regression is applied to derive Depth (D) from Draft (T), as highlighted in Section 3.1 of your paper.

```python
# Features and target for D prediction (D ~ T)
X_dt = df[['T']]
y_dt = df['D']

# Split the data
X_train_dt, X_test_dt, y_train_dt, y_test_dt = train_test_split(X_dt, y_dt, test_size=0.25, random_state=42)

# Train Linear Regression model
model_dt = LinearRegression()
model_dt.fit(X_train_dt, y_train_dt)

# Make predictions
y_pred_dt = model_dt.predict(X_test_dt)

# Evaluate the model
mse_dt = mean_squared_error(y_test_dt, y_pred_dt)
rmse_dt = sqrt(mse_dt)
mae_dt = mean_absolute_error(y_test_dt, y_pred_dt)
r2_dt = r2_score(y_test_dt, y_pred_dt)

# Output regression formula and metrics
print("Regression Formula for D (derived from T):")
print(f"D = {model_dt.intercept_:.2f} + ({model_dt.coef_[0]:.2f}) * T\n")
print("Evaluation Results:")
print("MAE:", mae_dt)
print("MSE:", mse_dt)
print("RMSE:", rmse_dt)
print("RÂ² Score:", r2_dt)
```

# Regression 3 â€“ Predicting Displacement Using CatBoost

To capture the non-linear relationship for ship displacement (as detailed in Section 3.1), we use a CatBoost Regressor with features T, BWL, and LWL.

```python
# Features and target for Displacement prediction
X_disp = df[['T', 'Bwl', 'Lwl']]
y_disp = df['D']  # 'D' is used as displacement in this context

# Split the data
X_train_disp, X_test_disp, y_train_disp, y_test_disp = train_test_split(X_disp, y_disp, test_size=0.25, random_state=42)

# Initialize and train the CatBoost Regressor
model_disp = CatBoostRegressor(iterations=500, learning_rate=0.1, depth=6, verbose=100)
model_disp.fit(X_train_disp, y_train_disp)

# Make predictions
y_pred_disp = model_disp.predict(X_test_disp)

# Evaluate the model
mse_disp = mean_squared_error(y_test_disp, y_pred_disp)
rmse_disp = sqrt(mse_disp)
mae_disp = mean_absolute_error(y_test_disp, y_pred_disp)
r2_disp = r2_score(y_test_disp, y_pred_disp)

# Custom rounding (as per research paper reporting)
def custom_round(x):
    return np.floor(x + 0.5)

print("CatBoost Regression (Displacement Prediction) Evaluation:")
print("MAE:", custom_round(mae_disp))
print("MSE:", custom_round(mse_disp))
print("RMSE:", custom_round(rmse_disp))
print("RÂ² Score:", custom_round(r2_disp * 100), "%")

# Plot Actual vs. Predicted Displacement
plt.figure(figsize=(10, 6), dpi=300)
plt.scatter(y_test_disp, y_pred_disp, color='red', marker='x', label='Predicted vs Actual')
plt.plot([min(y_test_disp), max(y_test_disp)], [min(y_test_disp), max(y_test_disp)], color='blue', linestyle='--', label='Perfect Prediction')
plt.xlabel('Actual Displacement (D)', fontsize=12)
plt.ylabel('Predicted Displacement (Å·)', fontsize=12)
plt.title('CatBoost Regression: Actual vs Predicted Displacement', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

# Model Training & Evaluation for Resistance Prediction (Rr)

This cell uses GridSearchCV to optimize several machine learning models 
(RandomForest, XGBoost, LightGBM, CatBoost, SVM, AdaBoost) with an expanded 
hyperparameter grid. It performs 10-Fold Cross-Validation to evaluate each model, 
calculates error metrics (MAE, MSE, RMSE, RÂ²), and visualizes the results.
The approach mirrors the detailed model comparison and performance assessment 
described in the research paper.

```python
# --- Load Data ---
df = pd.read_csv(r'C:\Users\poory\Desktop\New folder\yacht_hydro.csv')

# Define Features and Target (target is Rr, representing resistance)
X = df.drop(columns=['Rr'])
y = df['Rr']

# Setup 10-Fold Cross-Validation as described in the paper
cv = KFold(n_splits=10, shuffle=True, random_state=42)

# --- Define Models with Expanded Hyperparameter Grids ---
models = {
    'RandomForest': {
        'model': RandomForestRegressor(random_state=42),
        'params': {
            'n_estimators': [100, 200, 300],
            'max_depth': [None, 10, 20, 30],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    },
    'XGBoost': {
        'model': XGBRegressor(random_state=42, verbosity=0),
        'params': {
            'n_estimators': [100, 200, 300],
            'learning_rate': [0.01, 0.05, 0.1],
            'max_depth': [3, 5, 7],
            'subsample': [0.8, 1],
            'colsample_bytree': [0.8, 1]
        }
    },
    'LightGBM': {
        'model': LGBMRegressor(random_state=42),
        'params': {
            'n_estimators': [100, 200, 300],
            'learning_rate': [0.01, 0.05, 0.1],
            'max_depth': [-1, 10, 20],
            'num_leaves': [31, 50, 70]
        }
    },
    'CatBoost': {
        'model': CatBoostRegressor(random_state=42, verbose=0),
        'params': {
            'iterations': [100, 200, 300],
            'learning_rate': [0.01, 0.05, 0.1],
            'depth': [4, 6, 8],
            'l2_leaf_reg': [1, 3, 5]
        }
    },
    'SVM': {
        'model': SVR(),
        'params': {
            'C': [0.1, 1, 10, 100],
            'kernel': ['linear', 'rbf', 'poly'],
            'gamma': ['scale', 'auto'],
            'epsilon': [0.01, 0.1, 0.5]
        }
    },
    'AdaBoost': {
        'model': AdaBoostRegressor(random_state=42),
        'params': {
            'n_estimators': [50, 100, 200, 300],
            'learning_rate': [0.01, 0.05, 0.1, 0.5, 1],
            'loss': ['linear', 'square', 'exponential']
        }
    }
}

# --- Train and Evaluate Models with 10-Fold Cross-Validation ---
results = []
predictions = {}  # To store cross-validated predictions for each model

for name, config in models.items():
    print(f"\nTraining {name}...")
    grid = GridSearchCV(config['model'], config['params'], cv=cv, 
                        scoring='neg_mean_squared_error', n_jobs=-1)
    grid.fit(X, y)
    
    best_model = grid.best_estimator_
    y_pred = cross_val_predict(best_model, X, y, cv=cv)
    
    mae = mean_absolute_error(y, y_pred)
    mse = mean_squared_error(y, y_pred)
    rmse = sqrt(mse)
    r2 = r2_score(y, y_pred)
    
    results.append({
        'Model': name,
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'R2': r2,
        'Best Params': grid.best_params_
    })
    
    predictions[name] = y_pred

# --- Display Results in Table ---
results_df = pd.DataFrame(results)
print("\nModel Comparison:\n")
print(results_df.sort_values(by='R2', ascending=False))

# --- Plot 1: Bar Charts of Evaluation Metrics ---
metrics = ['MAE', 'RMSE', 'R2']
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for i, metric in enumerate(metrics):
    axes[i].bar(results_df['Model'], results_df[metric], color='skyblue')
    axes[i].set_title(metric)
    axes[i].set_xlabel('Model')
    axes[i].set_ylabel(metric)
    axes[i].set_xticklabels(results_df['Model'], rotation=45)
plt.suptitle('Comparison of Evaluation Metrics Across Models', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# --- Plot 2: Scatter Plots of Actual vs. Predicted for Each Model ---
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()
for i, (name, y_pred) in enumerate(predictions.items()):
    axes[i].scatter(y, y_pred, color='red', marker='x', alpha=0.6, label='Predicted vs Actual')
    axes[i].plot([y.min(), y.max()], [y.min(), y.max()], color='blue', linestyle='--', label='Perfect Prediction')
    axes[i].set_title(name)
    axes[i].set_xlabel('Actual Rr')
    axes[i].set_ylabel('Predicted Rr')
    axes[i].legend()
plt.suptitle('Actual vs. Predicted Rr for All Models', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
